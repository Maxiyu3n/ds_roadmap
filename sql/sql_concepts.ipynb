```python
# 5. what are window function, and how do they differ from aggregate functions?
```

Window functions perform calculations across a set of related rows, but unlike aggregate functions, they do not collapse rows. Instead, they return a value for each row, while still having access to data from a defined "window" or group.

Use aggregate functions when you want group-level summaries.

Use window functions when you need row-level data with group-level context.
